```python
import numpy as np 
import matplotlib.pyplot as plt
import h5py
from copy import deepcopy
params = {'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large'}
plt.rcParams.update(params)
```

```python
lams=np.array([1,2,4,5,8,10,20,40])
diffs=np.empty(0)
for lam in lams:
    opt=np.loadtxt(f"lam{lam}/flips.csv")
    diff=opt[0::2]-opt[1::2]
    diffs=np.append(diffs,diff.mean()/opt[0::2].mean())
print(diffs)
plt.figure(dpi=100)
plt.grid()
plt.xlabel("$\lambda/R_e$")
plt.ylabel("$\Delta_{flips}$")
plt.scatter(lams*0.125,diffs)
plt.savefig("dflips-lam.pdf")
```
