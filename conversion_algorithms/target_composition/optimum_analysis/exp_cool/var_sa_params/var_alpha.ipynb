```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Nbars=[10]
kappa=30/32
n_cells=40*40
V=5*5
V_cell=(1/8)**2
N=32
plt.figure(dpi=150)
plt.grid()
plt.xlabel(r"$\alpha$")
plt.ylabel("opt")
#plt.yscale('log')
for Nbar in Nbars:
    rho0=N*Nbar
    rhoc=rho0*V_cell #beads per cell
    var_phi=1/(rho0*kappa*(1/8)**2) 
    opt=np.loadtxt(f"Nbar{Nbar}/opt.csv")/(2*n_cells)
    alpha=np.loadtxt(f"Nbar{Nbar}/alpha.csv")
    n_alpha=len(alpha)#number of temperatures
    opt=opt.reshape(int(opt.shape[0]/n_alpha),n_alpha).mean(axis=0)
    plt.plot(alpha,opt/var_phi,lw=1,label=r"$\rho_{cell}=$"+f"{rhoc}")
#acc_rate=np.loadtxt(f"large_cells/acc_rate.csv")
#acc_rate=acc_rate.reshape(int(acc_rate.shape[0]/n_T),n_T).mean(axis=0)
#plt.plot(T,acc_rate,lw=1)
plt.legend()
plt.savefig("alpha.pdf")
print(320*(1/8)**2)
```
