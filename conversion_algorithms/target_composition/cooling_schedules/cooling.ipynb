```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

ks=np.empty(0)
Ts=np.empty(0)
T0=1.0
Tf=0.01
alpha=1/30*T0/Tf
T=T0
k=0
ks=np.append(ks,k)
Ts=np.append(Ts,T)
while(T>Tf):
    k+=1
    T=T0/(1+alpha*k)
    ks=np.append(ks,k)
    Ts=np.append(Ts,T)
    

plt.plot(ks,Ts)
print(k)
```
