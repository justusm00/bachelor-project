```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Simulations are run only with flips at $T=0$, e.g. no simulated annealing. Analyse runtime of one MC step and optimum of loss function as a function of system size. $\sqrt{\bar{N}}=200$ is kept constant while the box dimensions in $x$ and $z$ directions are varied, with $l_y=1R_e$ and polymer numbers are varied at constant discretization of $\frac{1}{8}R_e$ and lamella periodicity of $1R_e$. The polytypes are optimized every 1000 MC steps, 10 optimizations per system size.

```python
n_vals = 10 #time and opt values for each system size
time=np.loadtxt("time.csv")
opt=np.loadtxt("opt.csv")
lx=np.loadtxt("length.csv") #x dimension = z dimension
num_sim=int(len(lx)/n_vals) #number of simulations
time = time.reshape(num_sim,n_vals).mean(axis=1)
lx=lx.reshape(num_sim,n_vals).mean(axis=1)
opt = opt.reshape(num_sim,n_vals)

fix, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].set_title("Runtime")
axs[0].set_xlabel("Box dimensions in $R_e$")
axs[0].set_ylabel("Time in $s$")
axs[1].set_xlabel("Box dimensions in $R_e$")
axs[1].set_ylabel("Optimum of loss function")
axs[0].scatter(lx,time)
axs[1].scatter(lx,opt.mean(axis=1))
print(opt.mean(axis=1))
print(opt.std(axis=1))
```
