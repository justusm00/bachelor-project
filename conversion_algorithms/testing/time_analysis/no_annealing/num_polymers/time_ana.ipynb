```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Simulations are run only with flips at $T=0$, e.g. no simulated annealing. Analyse runtime of one MC step and optimum of loss function as a function of the number of polymers. The values are averaged over 50 different randomly generated configurations. Lamella periodicity is $\frac{1}{8}R_e$.

```python
n_vals = 100 #time and opt values for each number of polymers
time=np.loadtxt("poly_time.csv")
opt=np.loadtxt("poly_opt.csv")
n_poly=np.loadtxt("poly_num.csv")
num_sim=int(len(n_poly)/n_vals) #number of simulations
time = time.reshape(num_sim,n_vals).mean(axis=1)
n_poly=n_poly.reshape(num_sim,n_vals).mean(axis=1)
opt = opt.reshape(num_sim,n_vals)

fix, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].set_title("Runtime")
axs[0].set_xlabel("Number of polymers")
axs[0].set_ylabel("Time in $s$")
axs[1].set_xlabel("Number of polymers")
axs[1].set_ylabel("Optimum of loss function")
axs[0].scatter(n_poly,time)
axs[1].scatter(n_poly,opt.mean(axis=1))
print(opt.mean(axis=1))
print(opt.std(axis=1))
```
