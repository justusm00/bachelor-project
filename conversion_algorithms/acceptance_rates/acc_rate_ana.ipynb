```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Nbars=[10,50,100,200,500]
n_cells=40*40
N=32
plt.figure(dpi=150)
plt.grid()
plt.xlabel("$T$")
plt.ylabel("$r_{acc}$")
#plt.yscale('log')
plt.xscale('log')
for Nbar in Nbars:
    rhoc=N*Nbar/n_cells #beads per cell
    T=np.loadtxt(f"Nbar{Nbar}/T.csv")
    n_T=len(T)#number of temperatures
    acc_rate=np.loadtxt(f"Nbar{Nbar}/acc_rate.csv")
    acc_rate=acc_rate.reshape(int(acc_rate.shape[0]/n_T),n_T).mean(axis=0)
    plt.plot(T,acc_rate,lw=1,label=r"$\rho_{cell}=$"+f"{rhoc}")
#acc_rate=np.loadtxt(f"large_cells/acc_rate.csv")
#acc_rate=acc_rate.reshape(int(acc_rate.shape[0]/n_T),n_T).mean(axis=0)
#plt.plot(T,acc_rate,lw=1)
plt.legend()
plt.savefig("accrate.pdf")
print(20*(1/8)**3*5000/25*0.08/32)
```
