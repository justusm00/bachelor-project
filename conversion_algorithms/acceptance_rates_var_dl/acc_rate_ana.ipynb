#### Acceptance rates of SA algorithm for constant $\rho_{cell}=4.0$ and different $\sqrt{\bar N}$

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Nbars=[50,200,800]
dxs=[1/4,1/8,1/16]

plt.figure(dpi=150)
plt.grid()
plt.title(r"$\rho_c=4.0$")
plt.xlabel("$T$")
plt.ylabel("$r_{acc}$")
#plt.yscale('log')
plt.xscale('log')
for Nbar, dx in zip(Nbars,dxs):
    T=np.loadtxt(f"Nbar{Nbar}/T.csv")
    n_T=len(T)#number of temperatures
    acc_rate=np.loadtxt(f"Nbar{Nbar}/acc_rate.csv")
    acc_rate=acc_rate.reshape(int(acc_rate.shape[0]/n_T),n_T).mean(axis=0)
    plt.plot(T,acc_rate,lw=1,label=r"$\Delta L=$"+f"{dx}")
#acc_rate=np.loadtxt(f"large_cells/acc_rate.csv")
#acc_rate=acc_rate.reshape(int(acc_rate.shape[0]/n_T),n_T).mean(axis=0)
#plt.plot(T,acc_rate,lw=1)
plt.legend()
```
