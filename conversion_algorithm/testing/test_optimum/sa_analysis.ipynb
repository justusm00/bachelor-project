```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
lam_per=np.arange(1,10,1)*2/8
opt=np.array([])
for i in range(1,10):
    opt=np.append(opt,np.mean(np.loadtxt(f"opt_{i}.txt")))
print(opt)
plt.xlabel("Lamella periodicity [$R_e$]")
plt.ylabel("Optimum of loss function")
plt.grid()
plt.plot(lam_per,opt)
plt.hlines(0.014,np.min(lam_per),np.max(lam_per),color='g',label="thermal fluctuations")
plt.legend(loc="upper right")
plt.savefig("optimum_periodicity.pdf")
```
