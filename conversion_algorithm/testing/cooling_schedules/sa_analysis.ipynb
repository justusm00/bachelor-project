```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
T=np.loadtxt("T.txt")
acc_rate=np.loadtxt("acc_rate.txt")
mse=np.loadtxt("mse.txt")
```

```python
fig,axs=plt.subplots(2,figsize=(8,8))
axs[0].grid()
axs[1].grid()
axs[0].plot(T,acc_rate,lw=1)
axs[0].set_title("Acceptance rate")
axs[1].plot(T,mse,lw=1)
axs[0].set_xscale('log')
axs[1].set_xscale('log')
#axs[0].set_yscale('log')
#axs[1].set_yscale('log')
axs[1].set_title("MSE")
axs[1].set_xlabel("$T$")
fig.savefig("cooling_quad.pdf")
```
