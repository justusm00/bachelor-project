#### Script to analyse optimum of loss function for $\lambda=1/8R_e$ and $\sqrt{\bar N}=100$ (exponential SA cooling). All values are averaged over 100 or 1000 random initial configurations. Target densities alternate between 0.2 and 0.8.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

#### Constant parameters $T_{max}=10.0$ and $\alpha=0.85$

```python
opt_sa=np.loadtxt(f"sa/const_params/opt.csv")
print(f"Optimum with simulated annealing : {opt_sa.mean()}")
opt_no_sa=np.loadtxt("no_sa/opt.csv")
print(f"Optimum without simulated annealing : {opt_no_sa.mean()}")
print((opt_sa.mean()/opt_no_sa.mean()))
```

#### Variation of $T_{max}$ for $\alpha=0.85$

```python
tmax=np.loadtxt("sa/var_tmax/tmax.csv")
opt_no_sa=np.loadtxt("no_sa/opt.csv")
print(f"Optimum without simulated annealing: {opt_no_sa.mean()}")
n_sim=1000
opt_sa=np.loadtxt(f"sa/var_tmax/opt.csv")[:]
opt_sa=opt_sa.reshape(int(len(opt_sa)/n_sim),n_sim)
print(opt_sa.mean(axis=1))
plt.figure(dpi=100)
plt.grid()
plt.xlabel("$T_{max}$")
plt.ylabel("$L$")
plt.scatter(tmax,opt_sa.mean(axis=1))
```

#### Variation of $\alpha$ for $T_{max}=5.0$

```python
alpha=np.loadtxt("sa/var_alpha/alpha.csv")
opt_no_sa=np.loadtxt("no_sa/opt.csv")
print(f"Optimum without simulated annealing: {opt_no_sa.mean()}")
n_sim=100
opt_sa=np.loadtxt(f"sa/var_alpha/opt.csv")[:]
print(opt_sa.shape)
opt_sa=opt_sa.reshape(int(len(opt_sa)/n_sim),n_sim)
print(opt_sa.mean(axis=1))
plt.figure(dpi=100)
plt.grid()
plt.xlabel(r"$\alpha$")
plt.ylabel("$L$")
plt.scatter(alpha,opt_sa.mean(axis=1))
```
