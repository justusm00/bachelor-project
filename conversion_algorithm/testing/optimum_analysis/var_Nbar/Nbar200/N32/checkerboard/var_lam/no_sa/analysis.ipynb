```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
n_sim=10 #number of simulations per lambda
lam=np.loadtxt("lam.csv")*(1/4)
opt=np.loadtxt("opt.csv")
opt=opt.reshape(int(len(opt)/n_sim),n_sim).mean(axis=1)
print(lam)
print(opt)
plt.xlabel("$\lambda$ $[R_e]$")
plt.ylabel("$\Delta$")
plt.grid()
plt.scatter(lam,opt)
plt.savefig("optimum_periodicity.pdf")
```
